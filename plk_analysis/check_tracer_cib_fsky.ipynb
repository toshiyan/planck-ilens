```python
import numpy as np, plottools as pl, healpy as hp, local, tools_cib, binning as bn, curvedsky as cs, pickle, warnings
from matplotlib.pyplot import *
warnings.filterwarnings("ignore")
```

```python
iobj40 = tools_cib.init_cib(wind='G40')
iobj60 = tools_cib.init_cib(wind='G60')
```

```python
L, II40 = np.loadtxt(iobj40.fcII,usecols=(0,1)).T
L, II60 = np.loadtxt(iobj60.fcII,usecols=(0,1)).T
```

```python
pl.plot_1dstyle(ymin=1e-13,ymax=1e-11,ylog=True)
plot(L,II40)
plot(L,II60)
```

```python
pl.plot_1dstyle(ymin=1e-12,ymax=1e-9,xmin=1,xmax=100,ylog=True)
plot(L,II40)
plot(L,II60)
```
