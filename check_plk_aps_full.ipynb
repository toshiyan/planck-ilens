```python
import numpy as np, plottools as pl, healpy as hp, local, tools_cmb, binning as bn
from matplotlib.pyplot import *
import warnings
warnings.filterwarnings("ignore")
```

```python
aobj = local.init_analysis(freq='143',dtype='full',wind='Lmask')
```

```python
mb = bn.multipole_binning(50,lmin=500,lmax=2000)
```

```python
for cn in range(1,7):
    mcb, vcb, scb, ocb = bn.binned_spec(mb,aobj.fcmb.cl['c'],cn=cn,lfac=aobj.lfac)
    mnb, vnb, snb = bn.binned_spec(mb,aobj.fcmb.cl['n'],cn=cn,lfac=aobj.lfac,doreal=False)
    pl.plot_1dstyle(xmin=mb.lmin,xmax=mb.lmax)
    errorbar(mb.bc,ocb-mnb,yerr=vcb,fmt='o')
    plot(mb.bc,mcb-mnb)
    if cn<=4: plot(aobj.l,aobj.lfac*aobj.lcl[cn-1,:])
    show()
```
