```python
import numpy as np, plottools as pl, healpy as hp, local, tools_cmb, binning as bn
from matplotlib.pyplot import *
import warnings
warnings.filterwarnings("ignore")
```

```python
aobj = local.init_analysis(freq='143',dtype='full',wind='Lmask')
```

```python
mb = bn.multipole_binning(50,lmin=500,lmax=2000)
```

```python
mEB, vEB, __, oEB = bn.binned_spec(mb,aobj.fcmb.cl['c'],cn=6,lfac=aobj.lfac)
nEB, __, __ = bn.binned_spec(mb,aobj.fcmb.cl['n'],cn=6,lfac=aobj.lfac,doreal=False)
```

```python
mEE, vEE, __, oEE = bn.binned_spec(mb,aobj.fcmb.cl['c'],cn=2,lfac=aobj.lfac)
nEE, __, __ = bn.binned_spec(mb,aobj.fcmb.cl['n'],cn=2,lfac=aobj.lfac,doreal=False)
mBB, vBB, __, oBB = bn.binned_spec(mb,aobj.fcmb.cl['c'],cn=3,lfac=aobj.lfac)
nBB, __, __ = bn.binned_spec(mb,aobj.fcmb.cl['n'],cn=3,lfac=aobj.lfac,doreal=False)
```

```python
beta = .35*np.pi/180.
#beta = 0.
pl.plot_1dstyle(xmin=mb.lmin,xmax=mb.lmax,ymin=-5e-13,ymax=5e-13)
errorbar(mb.bc,oEB-(oEE-oBB)*beta*2.,yerr=vEB,fmt='o')
plot(mb.bc,mEB)
#plot(mb.bc,mcb-mnb)
```
