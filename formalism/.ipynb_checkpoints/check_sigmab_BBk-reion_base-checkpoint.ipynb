```python
import numpy as np, curvedsky as cs, plottools as pl, cmb as CMB, local, basic, tqdm
from matplotlib.pyplot import *
cr = rcParams['axes.prop_cycle'].by_key()['color']
```

```python
def snr_ilens(Ag,ucl,L,corr=1.):
    SN = (2*L+1.)*ucl[3]/Ag * corr**2
    sn = np.pi/180. * np.sqrt( np.sum(SN[2:]) )
    return sn
```

```python
def snr_eb(lcl,ocl,L,Lmin,Lmax):
    SN = (2*L[Lmin:Lmax+1]+1.)*(lcl[1,Lmin:Lmax+1]-lcl[2,Lmin:Lmax+1])**2/( ocl[1,Lmin:Lmax+1]*ocl[2,Lmin:Lmax+1] )
    sn = np.pi/180. * 2. * np.sqrt( np.sum(SN) ) 
    return sn
```

```python
Lmax  = 4000       # maximum multipole of output normalization
L = np.linspace(0,Lmax,Lmax+1)
lmin, lmax = 2, 10
rlmin, rlmax = 2, 4000
```

```python
ucl = CMB.read_camb_cls('../data_local/cosmo2017_10K_acc3_scalCls.dat',output='array')[:,:Lmax+1]
lcl = CMB.read_camb_cls('../data_local/cosmo2017_10K_acc3_lensedCls.dat',ftype='lens',output='array')[:,:Lmax+1]
```

```python
sigmas = np.logspace(-1,1,10)
```

```python
ClS = local.add_noise(lcl,2.,30.,lBmin=2)
```

```python
rcl = lcl.copy()
rcl[:,lmax+1:] = 0.
```

```python
sigma_eb = 1./snr_eb(rcl,ClS,L,lmin,lmax)
```

```python
theta = .5
#sigmas = np.array([1.])
```

```python
sigma_Amp  = np.zeros(len(sigmas))
sigma_Kinp = np.zeros(len(sigmas))
for i, sig in enumerate(sigmas):
    print('sigmaP=',sig)
    ClG = local.add_noise(lcl,sig,theta,lBmin=2,alpha=0.,fg=True)
    Ag = cs.norm_imag.qbb('amp',rlmax,rlmin,rlmax,rcl[1,:rlmax+1],ClG[2,:rlmax+1])[0]
    sigma_Amp[i]  = 1./snr_ilens(Ag,ucl[:,:rlmax+1],L[:rlmax+1],corr=1.)
    Ag = cs.norm_imag.qbb('lens',rlmax,rlmin,rlmax,rcl[1,:rlmax+1],ClG[2,:rlmax+1])[0]
    sigma_Kinp[i] = 1./snr_ilens(Ag,ucl[:,:rlmax+1],L[:rlmax+1],corr=1.)
    print(sigma_Kinp[i],sigma_Amp[i])
```

```python
sigma_amp  = np.zeros(len(sigmas))
sigma_krec = np.zeros(len(sigmas))
sigma_kinp = np.zeros(len(sigmas))
alpha      = np.zeros((len(sigmas),rlmax+1))
for i, sig in enumerate(sigmas):
    print('sigmaP=',sig)
    ClG = local.add_noise(lcl,sig,theta,lBmin=100,alpha=1.,fg=False)
    Ag = cs.norm_quad.qeb_iter(Lmax,rlmax,rlmin,rlmax,rlmin,rlmax,lcl[1,:],ClG[1,:],ClG[2,:],ucl[3,:],iter=50)[0]
    WE = np.ones((rlmax+1))
    Wp = ucl[3,:]/(ucl[3,:]+Ag)
    rho = ucl[3,:]/np.sqrt((ucl[3,:]+Ag)*ucl[3,:])
    LT = basic.delens.lintemplate(rlmax, rlmin, rlmax, rlmin, rlmax, lcl[1,:rlmax+1], ucl[3,:rlmax+1], WE, Wp[:rlmax+1])
    bb = basic.delens.lensingbb(rlmax, rlmin, rlmax, lcl[1,:rlmax+1], ucl[3,:rlmax+1])
    alpha[i] = 1-LT/bb
    ClG = local.add_noise(lcl,sig,theta,lBmin=2,alpha=alpha[i],fg=True)
    #ClG = local.add_noise(lcl,sig,theta,lBmin=100,alpha=alpha[i],fg=False)
    NL_amp = local.rec(rlmax,rlmin,rlmax,rcl[:,:rlmax+1],ClG[:,:rlmax+1],qdo=['BB'],est='amp')
    NL_len = local.rec(rlmax,rlmin,rlmax,rcl[:,:rlmax+1],ClG[:,:rlmax+1],qdo=['BB'],est='lens')
    #NL_amp = local.rec(rlmax,rlmin,rlmax,rcl[:,:rlmax+1],ClG[:,:rlmax+1],qdo=['BBt'],ocl1=ClS[:,:rlmax+1])
    #NL_len = local.rec(rlmax,rlmin,rlmax,rcl[:,:rlmax+1],ClG[:,:rlmax+1],qdo=['BBa'],ocl1=ClS[:,:rlmax+1])
    sigma_amp[i]  = 1./snr_ilens(NL_amp,ucl[:,:rlmax+1],L[:rlmax+1],corr=1.)
    sigma_kinp[i] = 1./snr_ilens(NL_len,ucl[:,:rlmax+1],L[:rlmax+1],corr=1.)
    sigma_krec[i] = 1./snr_ilens(NL_len,ucl[:,:rlmax+1],L[:rlmax+1],corr=rho[:rlmax+1])
    print(sigma_eb,sigma_krec[i],sigma_kinp[i])
```

```python
for i, sig in enumerate(sigmas):
    loglog(alpha[i])
```

```python
pl.plot_1dstyle(usetex=True)
xlabel(r'$\sigma_P$ [$\mu$K-arcmin]')
ylabel(r'$\sigma_\beta\sqrt{f_{\rm sky}}$ [deg]')
loglog(sigmas,sigma_krec,label=r'$\kappa$',color=cr[0])
loglog(sigmas,sigma_kinp,label=r'$\kappa$ (+ ideal tracer)',color=cr[1])
loglog(sigmas,sigma_amp*1e-1,label=r'$\delta\tau$ ($\sigma_\beta/10$)',color=cr[2])
loglog(sigmas,sigma_Kinp,ls='--',color=cr[1])
loglog(sigmas,sigma_Amp*1e-1,ls='--',color=cr[2])
legend()
savefig('fig_const_reion_t'+str(theta).replace('.','p')+'_lmax'+str(rlmax)+'.pdf')
```
